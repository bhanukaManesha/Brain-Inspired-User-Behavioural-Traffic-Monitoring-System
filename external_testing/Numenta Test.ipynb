# Testing the numenta model

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
from matplotlib.pyplot import figure
figure(num=None, figsize=(50, 50), dpi=80, facecolor='w', edgecolor='k')
```

```python
df_real = pd.read_csv('data/one.csv')
df_anomaly = pd.read_csv('data/one_out.csv')
```

```python
df_real.columns = ['timestamp','total','total_tcp','total_http','total_udp','size','size_tcp','size_http','size_udp']
df_real.head()
# df_real.dtypes
```

```python
df_anomaly.columns = ['timestamp','total','total_tcp','total_http','total_udp','size','size_tcp','size_http','size_udp']
df_anomaly.head()
# df_anomaly.dtypes
```

```python
ax1 = df_real.plot(y='total',x='timestamp',figsize=(100,20), title='Timestamp vs Total')
ax2 = df_anomaly.plot(y='total',x='timestamp',figsize=(100,20),color='red', title='Timestamp vs Total_Anomaly_Score')

ax2.set_ylim(0.9,1)
ax1.set_ylabel("Total")
ax2.set_ylabel("Total_Anomaly_Score")
```
